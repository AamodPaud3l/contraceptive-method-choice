### Contraceptive Method Choice Dataset

### Introduction

In this project, I have picked one dataset from the pool of datasets from the UCI repository at https://archive.ics.uci.edu/ml/datasets.php, the 'Contraceptive Method Choice' Data Set.  

Attribute Information:

1. Wife's age (numerical)
2. Wife's education (categorical) 1=low, 2, 3, 4=high
3. Husband's education (categorical) 1=low, 2, 3, 4=high
4. Number of children ever born (numerical)
5. Wife's religion (binary) 0=Non-Islam, 1=Islam
6. Wife's now working? (binary) 0=Yes, 1=No
7. Husband's occupation (categorical) 1=high level, 2, 3, 4=low level
8. Standard-of-living index (categorical) 1=low, 2, 3, 4=high
9. Media exposure (binary) 0=Good, 1=Not good
10. Contraceptive method used (class attribute) 1=No-use, 2=Long-term, 3=Short-term

## Data Modelling

```python
# Importing necessary library for train test split

from sklearn.model_selection import train_test_split
```

```python
# Separating the dependent and independent variables
X=df.drop('Contraceptive method used',axis=1)
y=df['Contraceptive method used']
```

### Splitting the data

```python
# Splitting the data

#Suite 1
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.50, random_state=0)

#Suite 2
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, test_size=0.40, random_state=0)

#Suite 3
X_train3, X_test3, y_train3, y_test3 = train_test_split(X, y, test_size=0.20, random_state=0)
```

### Performing Data Modelling

### KNN Classifier -3 suites

```python
# Importing KNN Classifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn import metrics

x = range(1,50)
```

```python
# Suite 1
acc = []
final_acc = []
for i in x:
    knn_trained = KNeighborsClassifier(n_neighbors = i).fit(X_train1,y_train1)
    y_test_pred = knn_trained.predict(X_test1)
    acc.append(metrics.accuracy_score(y_test1, y_test_pred))
    
plt.figure(figsize=(10,6))
plt.plot(range(1,50),acc)
plt.title('accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Maximum accuracy:",max(acc),"at K =",acc.index(max(acc)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

knn_trained = KNeighborsClassifier(n_neighbors = 24).fit(X_train1,y_train1)
y_test_pred = knn_trained.predict(X_test1)

# Print results: 
print(confusion_matrix(y_test1, y_test_pred))
print(classification_report(y_test1, y_test_pred))

# final_acc to compare results
final_acc.append(('KNN1',metrics.accuracy_score(y_test1, y_test_pred)))
```

```python
# Suite 2

acc = []
for i in x:
    knn_trained = KNeighborsClassifier(n_neighbors = i).fit(X_train2,y_train2)
    y_test_pred = knn_trained.predict(X_test2)
    acc.append(metrics.accuracy_score(y_test2, y_test_pred))
    
plt.figure(figsize=(10,6))
plt.plot(range(1,50),acc)
plt.title('accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Maximum accuracy:",max(acc),"at K =",acc.index(max(acc)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

knn_trained = KNeighborsClassifier(n_neighbors = 24).fit(X_train2,y_train2)
y_test_pred = knn_trained.predict(X_test2)

# Print results: 
print(confusion_matrix(y_test2, y_test_pred))
print(classification_report(y_test2, y_test_pred))

final_acc.append(('KNN2',metrics.accuracy_score(y_test2, y_test_pred)))
```

```python
# Suite 3

acc = []
for i in x:
    knn_trained = KNeighborsClassifier(n_neighbors = i).fit(X_train3,y_train3)
    y_test_pred = knn_trained.predict(X_test3)
    acc.append(metrics.accuracy_score(y_test3, y_test_pred))
    
plt.figure(figsize=(10,6))
plt.plot(range(1,50),acc)
plt.title('accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Maximum accuracy:",max(acc),"at K =",acc.index(max(acc)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

knn_trained = KNeighborsClassifier(n_neighbors = 24).fit(X_train3,y_train3)
y_test_pred = knn_trained.predict(X_test3)

# Print results: 
print(confusion_matrix(y_test3, y_test_pred))
print(classification_report(y_test3, y_test_pred))

final_acc.append(('KNN3',metrics.accuracy_score(y_test3, y_test_pred)))
```

### Decision Tree Classifier - 3 suites

```python
# Importing Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier 

#Suite 1
dt = DecisionTreeClassifier()
dt_trained = dt.fit(X_train1, y_train1) # train set
y_pred = dt_trained.predict(X_test1) # predict target

print(confusion_matrix(y_test1, y_pred)) 

acc = metrics.accuracy_score(y_test1, y_pred) 
print(acc)

print(classification_report(y_test1, y_pred))
```

```python
# Using the stratified 5-fold cross-validation to identify the best value of the hyperparameter max_depth

from sklearn.model_selection import cross_val_score #Evaluate a score by cross-validation
import numpy as np
depths = range(1, 6)
print(depths)

cv_scores = []
for maxDpt in depths:
    dt = DecisionTreeClassifier(max_depth = maxDpt)
    scores = cross_val_score(dt, X_train1, y_train1, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())
    print(scores)
maxDpt_best = depths[np.argmax(cv_scores)] #Returns the indices of the maximum values along an axis.
print('cv_scores:', cv_scores)
print('maxDpt_best:', maxDpt_best)
```

```python
# Using the depth value = 4
dt = DecisionTreeClassifier(max_depth = 4)
dt_trained = dt.fit(X_train1, y_train1)
y_pred = dt_trained.predict(X_test1)

cm = confusion_matrix(y_test1, y_pred) 
print(cm)

acc = metrics.accuracy_score(y_test1, y_pred) 
print(acc)

print(classification_report(y_test1, y_pred))

final_acc.append(('DT1',acc))
```
